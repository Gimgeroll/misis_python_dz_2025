Задача 1: Частотный словарь (★★☆☆☆)
Цель: Попрактиковаться в работе со словарями и строковыми методами.

Описание: Напишите функцию count_word_frequency(text), которая принимает на вход строку text и возвращает словарь, где ключи — это уникальные слова из текста, а значения — количество раз, которое каждое слово встретилось.

Требования:

Функция должна приводить весь текст к нижнему регистру, чтобы "Слово" и "слово" считались одним и тем же.
Функция должна избавляться от знаков препинания в тексте (например, точек, запятых, восклицательных знаков). Простой способ — использовать replace().
Слова в тексте разделены пробелами.
Пример:

text = "Это просто пример. Просто очень простой пример текста для примера."
frequency_dict = count_word_frequency(text)
print(frequency_dict)
# Ожидаемый результат:
# {'это': 1, 'просто': 2, 'пример': 3, 'очень': 1, 'простой': 1, 'текста': 1, 'для': 1}
Подсказка:

Чтобы избавиться от знаков препинания, можно пройтись по списку [".", ",", "!", "?"] и для каждого знака вызывать text = text.replace(znak, "").
Используйте метод .split() для получения списка слов.
Используйте метод словаря .get(key, 0) для удобного подсчета.

```python
def count_word_frequency (text):
    """
    Функция ведёт подсчёт количества слов в тексте
    Args = текст, в котором надо подсчитать кол-во повторений
    return = словарь, в котором ключи - кол-во повторений слов, а значения - уникальные слова из текста
    """
    text = text.lower()
    punct_marks = [".", ",", "!", "?", ":", ";", "-", "—", "(", ")", "[", "]", "{", "}", "'", "\""]
    for mark in punct_marks:
        text = text.replace(mark, '')
        
    words = text.split()
    
    frequency_dict={}
    
    for word in words:
        frequency_dict[word] = frequency_dict.get(word, 0) + 1
    return frequency_dict

text = input('Введите текст')
print(count_word_frequency(text))
```

Цель: Попрактиковаться в написании функций с параметрами по умолчанию и именованными аргументами.

Описание: Напишите функцию calculate_price(price, discount=0, tax_rate=0.2). Функция должна рассчитывать итоговую стоимость товара.

Требования:

Функция принимает price (обязательный позиционный аргумент).
discount — скидка в процентах (необязательный, по умолчанию 0).
tax_rate — налоговая ставка в долях (необязательный, по умолчанию 0.2, что соответствует 20% НДС).
Сначала применяется скидка, а затем на цену со скидкой начисляется налог.
Функция должна возвращать итоговую стоимость.

```python
def calculate_price(price, discount=0, tax_rate=0.2):
    total_price = 0
    if price<=0:
        print('Ошибка: цена не может быть меньше нуля')
    disc_price = price*(1-discount/100)
    total_price = disc_price*(1+tax_rate)
    return total_price
    
print("Пример 1 (только цена):", calculate_price(1000))
print("Пример 2 (цена + скидка 10%):", calculate_price(1000, discount=10))
print("Пример 3 (цена + налог 10%):", calculate_price(1000, tax_rate=0.1))
print("Пример 4 (цена + скидка 15% + налог 15%):", calculate_price(1000, discount=15, tax_rate=0.15))
```

```python
from datetime import datetime

def analyze_log(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()
    except FileNotFoundError:
        return {}
    
    counts = {'INFO': 0, 'WARNING': 0, 'ERROR': 0}
    min_date = None
    max_date = None
    
    for line in lines:
        line = line.strip()
        if not line:
            continue
            
        try:
            parts = line.split(',', 2)
            if len(parts) < 3:
                continue
                
            date_str = parts[0].strip()
            level = parts[1].strip()
            message = parts[2].strip()
            
            log_date = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
            
            if level in counts:
                counts[level] += 1
            
            if min_date is None or log_date < min_date:
                min_date = log_date
            if max_date is None or log_date > max_date:
                max_date = log_date
                
        except (ValueError, IndexError):
            continue
    
    if min_date and max_date:
        time_range = (
            min_date.strftime("%Y-%m-%d %H:%M:%S"),
            max_date.strftime("%Y-%m-%d %H:%M:%S")
        )
    else:
        time_range = (None, None)
    
    return {
        'counts': counts,
        'time_range': time_range
    }

result = analyze_log('log.txt')
print(result)
```
