```python
N = float(input('Введите целое положительно число'))
if N <= 0 or N%1!=0:
    while N < 0 or N%1!=0:
        print('Вас попросили ввести целое положительно число')
        N = float(input('Ещё раз введите целое положительно число'))
N=int(N)
for i in range (1, N+1):
    if i%3==0 and i%5==0:
        print('FizzBuzz')
    elif i%3==0:
        print('Fizz')
    elif i%5==0:
        print('Buzz')
    else:
        print(i)
        
```

```python
shopping_list = []
while True:
    N = int(input('"1. Добавить", "2. Показать", "3. Выйти"'))
    if N == 1:
        pr = input('Введите название продукта: ')
        shopping_list.append(pr)
    elif N == 2:
        if len(shopping_list) == 0:
            print ('Список пуст')
        else:
            for index, prdct in enumerate(shopping_list):
                print(f"Продукт №{index+1}: {prdct}")
    elif N == 3:
        print('Всего хорошего!')
        break
    else:
        print('Чумба, ты ебанулся? Тебе дали три опции на выбор, не выдумывай')
```
